```python
import pandas as pd

data = pd.read_excel('mp1c00469_si_002.xlsx')
data
```

```python
data.info() #there is no missing data!
```

```python
data.describe() #high variability for NaCl
```

```python
data.describe(include=object)
```

```python
import numpy as np

np.unique(data.Round) #3 rounds + initial conditions
```

```python
import seaborn as sns
sns.set_theme()

#clearly see that the initial round is disperse and the last round is more concentrated (maybe a local minimun)
sns.jointplot(data=data, x="Tm", y="NP", hue="Round") 
```

```python
sns.relplot(
    data=data,
    x="Exp No.", y="NP", col="Round",
    hue="pH", size="NP_std"#, style="pH",
)
```

```python
sns.pairplot(data=data[['Round', 'pH', 'NaCl (mM)', 'Trehalose (w/v %)',
                        'Mannitol (w/v %)','Tween 20 (v/v%)','Tm','NP','NP_std']], hue="Round")
```

```python
g = sns.PairGrid(data, hue="Round", corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```
